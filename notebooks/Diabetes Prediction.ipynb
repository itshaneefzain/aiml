```python
# replace 0's with nan
cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

df[cols] = df[cols].replace(0, np.nan)
```

```python
# here we are filling with median() instead if mean() because medical data is skewd and insulin will have extreme outliers, 
# filling it with mean will distort the values

cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for i in cols:
    df[i] = df[i].fillna(df[i].median(numeric_only=True)) #numeric_only = True fill only the numeric value  here its not necessary as we
    # selected cols explicitly

df.head(10)
```
